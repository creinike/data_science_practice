## Metrics for Regression

```python
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

## Explained Variance Score

```python
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
print(f"Explained Variance Score: {explained_variance_score(y_true, y_pred)}")  
```

## Mean Absolute Error

```python
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
print(f"Mean Absolute Error: {mean_absolute_error(y_true, y_pred)}")
```

## Mean Squared Error

```python
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
print(f"Mean Squared Error: {mean_squared_error(y_true, y_pred)}")
```

## R2 Score

```python
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
print(f"R2 Score: {r2_score(y_true, y_pred)}") 
```

## Regularization

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
X = 1. / (np.arange(1,11) + np.arange(0,10)[:, np.newaxis]) # Hilbert matrix of 10x10
y = np.ones(10)
print(X)
```

```python
n_alphas = 100
alphas = np.logspace(-10, 2, n_alphas)

ridge_coefs = []

for a in alphas:
    ridge = linear_model.Ridge(alpha=a, fit_intercept=False)
    ridge.fit(X,y)
    ridge_coefs.append(ridge.coef_)
```

```python
print(ridge_coefs)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(alphas, ridge_coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1]) #reverse axis
ax.set_xlabel('alpha')
ax.set_ylabel('weights')
ax.set_title(f'Ridge coefficients as a function of regularization')
```
