```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

## <u>Scatter:</u>

```python
x = np.linspace(-5,5,20)
y = x**2
plt.scatter(x,y, s=200, color='yellow',marker='.',edgecolors='black')
```

## <u>Bar Chart:</u>

```python
classes = ['C1', 'C2', 'C3', 'C4']
colors  = ['#FF5733', '#3533FF', '#33FF5D', '#FDFF33']
numerical = [[10, 15, 5, 7],
             [1, 3, 4, 9],
             [4, 15, 8, 10],
             [7, 6, 13, 8]]

number_groups = len(classes) 
bin_width = 0.2

fig, ax = plt.subplots(figsize=(8,6))

for i in range(number_groups):
    ax.bar(x=np.arange(len(classes)) + i*bin_width, 
           height=numerical[i],
           width=bin_width,
           color=colors[i],
           align='center')

ax.set_xticks(np.arange(len(classes)) + 0.4)
ax.set_xticklabels(classes)
ax.legend(classes, facecolor='w', loc='best')
```

## <u>Histogram:</u>

```python
x = np.random.sample(50)
n, bins, patches = plt.hist(x, bins=20, facecolor='orange', alpha=0.7, density=True)
plt.setp(patches[3], 'facecolor', 'b')
plt.xlabel('bins')
plt.ylabel('Values')
plt.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
```

n: is the number of counts in each bin of the histogram

bins: is the left hand edge of each bin

patches: is the individual patches used to create the histogram, e.g a collection of rectangles

## <u>Pie Chart:</u>

```python
# Pie chart, where the slices are ordered and plotted counter-clockwise:
labels = 'Segment 1', 'Segment 2', 'Segment 3', 'Segment 4'
sizes = [30, 35, 25, 10]  # percentages, the sum of all of them must to be 100%
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Segment 2')

fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
```

## <u>Box Plot:</u>

* Q3: This is the 75th percentile value (upper hinge) of the data.
* Q3: This is the 25th percentile value (lower hinge) of the data.
* Box: It is a difference between the upper hinge and the lower hinge.
* Median: This is the midpoint of the data.
* Max: This is the upper inner fence. It is 1.5 times the step above Q3.
* Min: This is the lower inner fence. It is 1.5 times the step below Q1.
* Any value that is greater than Max or lesser than Min is called an outlier.
* approximately 95% or the data are expected to lie between the inner fences

```python
_high = np.random.rand(20) + 0.5
center = np.random.rand(20)
_low = np.random.rand(20) - 0.5

data = [_high,center,_low]
plt.boxplot(data,notch=True, vert=True, patch_artist=True)
```

## <u>Violin Plot:</u>

```python
plt.figure(figsize=(9,4))
plt.violinplot(data,widths=0.40, vert=True, showmeans=True, showextrema=True, showmedians=True)
```

```python
fig, axes = plt.subplots(figsize=(9, 4))
axes.violinplot(data, widths=0.40, vert=True, showmeans=True, showextrema=True, showmedians=True)
axes.set_title('Violinplot', fontsize=10)
plt.setp(axes, xticks=[i+1 for i in range(len(data))], xticklabels=['high','center','low'])
```

## <u>Bubble Charts:</u>

```python
plt.style.use('ggplot')
df = pd.DataFrame(np.random.rand(50, 3), columns=['x','y','z'])
df.plot(kind='scatter',x='x',y='y',s=df['z']*400, color='red', edgecolors='black')
# s is the size of the bubble
```

## <u>3D Plots:</u>

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
img = plt.figure()
ax = Axes3D(img)
x = np.arange(-5,5,0.25)
y = np.arange(-5,5,0.25)
X, Y = np.meshgrid(x, y)
V = np.sqrt(X**2 + Y**2)
Z = np.sin(V)
ax.plot_surface(X,Y,Z, rstride=2, cstride=2, cmap='hot',linewidth=0)
```

#### Color bar:

```python
p = ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap='hot',linewidth=0)
img.colorbar(p, shrink=0.5)
img
# The rstride and cstride parameters help in sizing the cell on the surface.
```

#### Setting the view angle:

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.view_init(elev=0., azim=0)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
```

```python
# setting the view at 50 degrees elevation and 30 degrees angle:
fig = plt.figure()
ax = Axes3D(fig)
ax.view_init(elev=50., azim=30)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
```

## 3D Line Plot:

```python
fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection='3d')

v = np.linspace(-5 * np.pi, 5 * np.pi, 100)
z = np.linspace(-3, 3, 100)
r = z**2 + 1
x = r * np.sin(v)
y = r * np.cos(v)
 
ax.plot(x, y, z, label='3D curve')
ax.legend(loc='best')
```
